```python
from transformers import pipeline
import pandas as pd

classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

categories = ["Login Issue","Payment Problem","Technical Bug","Account Management","Performance Issue","Billing Issue"]

tickets = [
    "I cannot log into my account",
    "Payment was deducted twice",
    "App crashes when opening settings",
    "How do I update my profile?",
    "The website is very slow"
]

results = []

for t in tickets:
    res = classifier(t, categories)
    labels = res["labels"][:3]
    scores = res["scores"][:3]

    output = "\n".join([f"{i+1}. {labels[i]} ({round(scores[i],2)})" for i in range(3)])

    results.append({
        "Ticket": t,
        "Top 3 Tags": output
    })

df = pd.DataFrame(results)
print(df.to_string(index=False))
```
